```python
# %%
import numpy as np
import matplotlib.pyplot as plt
from bqplot import pyplot as bqplt
from bqplot import LinearScale, Scatter
def target_function(x):
    return np.exp(-(x-0.1) ** 2) * np.sin(6 * np.pi * x ** (3 / 4)) ** 2
X = np.linspace(0, 1, 1000)
y = np.array([target_function(x) for x in X])

plt.plot(X, y)
plt.show()
# %%
# bqplt.plot(X,y)
# bqplt.show()
# %%
x_sc = LinearScale()
y_sc = LinearScale()
scatter_chart = Scatter(x=X, y=y, scales={'x': x_sc, 'y': y_sc})
scatter_chart
```
